4. Mengatasi missing value


```python
# Mengatasi missing value
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
        df[column].fillna(df[column].mean(), inplace=True)
```

```python
# cek ulang missing value
df.isna().sum()
```

5. Cek Duplikat Data

```python
# Mengecek apakah ada duplicate di seluruh kolom
cek_duplikat = df.duplicated().sum()
print(f"Jumlah data duplikat = {cek_duplikat}")
```

# Analisis Tambahan

```python
# Mengelompokkan data berdasarkan 'City' dan menjumlahkan 'Total Spend'
city_spend = df.groupby('City')['Total Spend'].sum()

# Mengurutkan hasil secara menurun dan mengambil 5 teratas
top_5_cities_spend = city_spend.nlargest(5)

# Menampilkan hasilnya
print("5 Kota dengan Total Pengeluaran Terbanyak:")
display(top_5_cities_spend)
```

```python
# Filter data untuk pelanggan yang membeli dalam 30 hari terakhir
recent_purchases_df = df[df['Days Since Last Purchase'] <= 30]

# Menghitung jumlah pelanggan berdasarkan Satisfaction Level dalam data yang difilter
satisfaction_counts_recent = recent_purchases_df['Satisfaction Level'].value_counts()

# Menghitung total pelanggan dalam data yang difilter
total_recent_customers = len(recent_purchases_df)

# Menghitung persentase untuk setiap Satisfaction Level dalam data yang difilter
satisfaction_percentages_recent = (satisfaction_counts_recent / total_recent_customers) * 100

# Menampilkan hasilnya
print("Persentase Satisfaction Level untuk pelanggan yang membeli dalam 30 hari terakhir:")
display(satisfaction_percentages_recent)
```

```python
# Mengelompokkan data berdasarkan 'Membership Type' dan menjumlahkan 'Total Spend'
membership_spend = df.groupby('Membership Type')['Total Spend'].sum()

# Menghitung total pengeluaran keseluruhan
total_overall_spend = df['Total Spend'].sum()

# Menghitung persentase total pengeluaran untuk setiap Membership Type
membership_spend_percentage = (membership_spend / total_overall_spend) * 100

# Mengurutkan hasil persentase secara menurun
sorted_membership_spend_percentage = membership_spend_percentage.sort_values(ascending=False)

# Menampilkan hasilnya
print("Persentase Total Pengeluaran berdasarkan Membership Type:")
display(sorted_membership_spend_percentage)
```

# Visualisasi

Visualisasi 5 kota dengan total spend terbanyak

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat bar plot
plt.figure(figsize=(8, 4))
sns.barplot(x=top_5_cities_spend.index, y=top_5_cities_spend.values, palette='viridis')
# Menambahkan judul dan label
plt.title('5 Kota dengan Total Spend Terbanyak')
plt.xlabel('Kota')
plt.ylabel('Total Spend',fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Visualisasi Satisfaction Level berdasarkan 30 hari terakhir penjualan

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat pie chart dengan palet warna yang berbeda
plt.figure(figsize=(8, 8))
plt.pie(satisfaction_percentages_recent, labels=satisfaction_percentages_recent.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2')) # Mengganti palet warna menjadi 'Set2'
plt.title('Persentase Satisfaction Level (Pembelian 30 Hari Terakhir)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Visualisasi Total Spend berdasarkan membership type

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat horizontal bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=sorted_membership_spend_percentage.values, y=sorted_membership_spend_percentage.index, palette='viridis')

# Menambahkan judul dan label
plt.title('Persentase Total Pengeluaran berdasarkan Membership Type', fontsize=14)
plt.xlabel('Persentase (%)')
plt.ylabel('Jenis Keanggotaan')

# Menambahkan nilai persentase pada bar
for index, value in enumerate(sorted_membership_spend_percentage.values):
    plt.text(value, index, f'{value:.1f}%')

plt.tight_layout()
plt.show()
```
